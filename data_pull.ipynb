```python
import pandas as pd
```

```python
df = pd.read_csv('MEC Shot Quality 2025-26 - CSV Pull.csv')
df.tail()
```

```python
dates = df['Date'].unique()
home_teams = df['Home_Team'].unique()

date_ht_lst = []
for i in dates:
    for j in home_teams:
        if len(df[(df['Date']==i) & (df['Home_Team']==j)]):
            date_ht_lst.append([i, j])
len(date_ht_lst)
```

```python
for i in range(len(date_ht_lst)):
    temp_df = df[(df['Date']==date_ht_lst[i][0]) & (df['Home_Team']==date_ht_lst[i][1])]
    date = temp_df['Date'].iloc[0]
    month, day, year = date.split('/')
    home_team = temp_df['Home_Team'].iloc[0]
    away_team = temp_df['Away_Team'].iloc[0]
    print(date, home_team, away_team, month, day, year)
    file_name = f"Game_{home_team}_{away_team}_{month}.{day}.{year}"
    print(file_name)
    temp_df.to_csv(f'server/raw_data/{file_name}.csv', index=False)
```
